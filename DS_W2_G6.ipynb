```python
#Jessica---------------------------------
#Como estábamos teniendo problemas al abrir el archivo, lo que hice fue subirlo directamente a la carpeta de trabajo para tenero de rápido acceso
#Sin embargo, como el documento era muy pesado, lo que hice fue convertirlo en zip, subirlo y las siguientes líneas de código son para abrir el archivo zip

#Abrir zip
from io import StringIO
from io import BytesIO
from zipfile import ZipFile
import urllib.request
import pandas as pd 

link = "https://github.com/SteeleP-J/DS-Project/raw/master/Auto.zip"
url = urllib.request.urlopen(link)


with ZipFile(BytesIO(url.read())) as my_zip_file:
    for contained_file in my_zip_file.namelist():
        fzip=my_zip_file.open(contained_file)
        data=fzip.read()

#Convertir el archivo en un data frame con pandas
s=str(data,'utf-8')
data = StringIO(s) 
df=pd.read_csv(data)
df.head()
```

```python
df['name'] = df['name'].str.replace(r"_"," ")
df['name'] = df['name'].str.replace(r"\W"," ")
df['model'] = df['model'].str.replace(r"\W"," ")
df['name'] = df['name'].str.replace(r"  "," ")
df['name'] = df['name'].str.replace(r"  "," ")
df['name'] = df['name'].str.upper()
df['seller'] = df['seller'].str.title()
df['offerType'] = df['offerType'].str.title()
df['abtest'] = df['abtest'].str.title()
df['vehicleType'] = df['vehicleType'].str.title()
df['gearbox'] = df['gearbox'].str.title()
df['model'] = df['model'].str.title()
df['fuelType'] = df['fuelType'].str.title()
df['brand'] = df['brand'].str.title()
df['notRepairedDamage'] = df['notRepairedDamage'].str.title()
df['dateCrawled'] = pd.to_datetime(df['dateCrawled'],yearfirst=True, errors='coerce')
df['dateCreated'] = pd.to_datetime(df['dateCreated'],yearfirst=True, errors='coerce')
df['lastSeen'] = pd.to_datetime(df['lastSeen'],yearfirst=True, errors='coerce')
df['notRepairedDamage'] = df['notRepairedDamage'].fillna('No')  

df['name'] = df['name'].str.strip()
df.head()
```
